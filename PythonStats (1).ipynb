### Now we are goinig to use scripy.stats for simple statistical test

```python
from scipy import stats
```

### 1-sample t-test: testing the value of a population mean
    scip.stats.ttest_1samp() tests if the pop mean of data is likely to be equal to a given value. It returns the T statistic and the p-value 

```python
stats.ttest_1samp(data['VIQ'],0)
```

### 2-sample t-test: testing for difference across populations 
    We have seen above that the mean VIQ in the male and female populations were different. To test if this is significant, we do a 2 sample t-test with scipy.stats.ttest_ind()

```python
female_viq = data[data['Gender'] == 'Female']['VIQ']
male_viq = data[data['Gender'] == 'Male']['VIQ']
stats.ttest_ind(female_viq, male_viq)   
```

### Paired tests: repeated measurements on the same indivdiauls 
    PIQ, VIQ, and FSIQ gives use 3 measurements. We can use a 2 sample test to compare two of them first.  

```python
stats.ttest_ind(data['FSIQ'], data['PIQ'])
```

The problem with this approach is that it forgets that there are links between observations: FSIQ and PIQ are measured on the same individuals. This the variance due to inter-subject variability is confounding and can be removed, using a "paired test" or "repeated measures test"

```python
stats.ttest_rel(data['FSIQ'], data['PIQ'])
```

This is equivalent to a 1-sample test on the difference 

```python
stats.ttest_1samp(data['FSIQ'] - data['PIQ'], 0)
```

T-test assume Gaussian errors. We can use a Wilcoxon signed-rank test. that removes this assumption

```python
stats.wilcoxon(data['FSIQ'], data['PIQ'])
```

### A simple linear regression
    y = x * coef + intercept + e
    where e is the observation noise. We will use the statsmodels module to:
    1. Fit a linear model. We will use tehe simplest strat, ordinary least squares (OLS)
    2. Test that coef is non zero
    First we generate simulated data according to the model. 

```python
import numpy as np
```

```python
x = np.linspace(-5, 5, 20)
np.random.seed(1)
y = -5 + 3*x + 4 * np.random.normal (size=x.shape)
data = pd.DataFrame({'x': x, 'y': y})
```

We can specify an OLD model and fit. Then we can inspect the various statisics derived from the fit

```python
from statsmodels.formula.api import ols
model = ols("y ~ x", data).fit()
```

```python
print(model.summary())
```

### Categorical Variables: comparing groups or multiple categories 
    We are going to go back to the original data. We can write a comparison between IQ of male and female using a linear model

```python
data = pd.read_csv('brain_size.csv', sep=';', na_values=".")
model = ols("VIQ ~ Gender + 1", data).fit()
print(model.summary())
```

Forcing categorical: the Gender variable is automatically detected as a categorical variable. An integer column can be forced to be treated as categorical. 

```python
model = ols('VIQ ~ C(Gender)', data).fit()
```

Link to t-tests between different FSIQ and PIQ
    To compare different types of IQ, we need to create a "long-form" table, listing IQs, where the type of IQ is indicated by a categorical variable:

```python
data_fisq = pd.DataFrame({'iq': data['FSIQ'], 'type': 'fsiq'})
data_piq = pd.DataFrame({'iq': data['PIQ'], 'type': 'piq'})
data_long = pd.concat((data_fisq, data_piq))
print(data_long)
model = ols("iq ~ type", data_long).fit()
print(model.summary())
```

We can see that we retrieve the same values for t-test and corresponding p-values for the effect of the type of IQs than the previous t-test:

```python
stats.ttest_ind(data['FSIQ'], data['PIQ']) 
```

### Multiple Regression: including multiple factors 
    Now we are looking at regressions with multiple factors. We are now looking at a new data set; Iris. We are going to read the new csv file, create a model, and print the summary. 

```python
data_iris = pd.read_csv('iris.csv')
model_iris = ols('sepal_width ~ name + petal_length', data_iris).fit()
print(model_iris.summary())
```

### Post-hoc hypothesis testing: Anova 
    If we wish to test if the petal length is different between veriscolor and virginica, after removing the effect of sepal width, we would want to use an ANOVA

```python
print(model_iris.f_test([0, 1, -1, 0])) 
```

### Seaborn for statistical exploration
    seaborn combines simple statistical fits with plotting on pandas dataframes. We are going to load a new set of data about wages and other information on 500 individuals. First was are going to import and set up this data so we can use it for analysis with seaborn. 

```python
import os
import urllib.request
if not os.path.exists('wages.txt'):
    urllib.request.urlretrieve('http://lib.stat.cmu.edu/datasets/CPS_85_Wages',
                       'wages.txt')
names = [
    'EDUCATION: Number of years of education',
    'SOUTH: 1=Person lives in South, 0=Person lives elsewhere',
    'SEX: 1=Female, 0=Male',
    'EXPERIENCE: Number of years of work experience',
    'UNION: 1=Union member, 0=Not union member',
    'WAGE: Wage (dollars per hour)',
    'AGE: years',
    'RACE: 1=Other, 2=Hispanic, 3=White',
    'OCCUPATION: 1=Management, 2=Sales, 3=Clerical, 4=Service, 5=Professional, 6=Other',
    'SECTOR: 0=Other, 1=Manufacturing, 2=Construction',
    'MARR: 0=Unmarried,  1=Married',
]

short_names = [n.split(':')[0] for n in names]

data = pd.read_csv('wages.txt', skiprows=27, skipfooter=6, sep=None,
                       header=None, engine='python')
data.columns = short_names
```

We can use seaborn.pairplot() to display a scatter matrix that can displat the interactions between continous variables 

```python
import seaborn
```

```python
seaborn.pairplot(data, vars=['WAGE','AGE','EDUCATION'], kind='reg')
```

categorical variables can be plotted as the hue

```python
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'], kind='reg', hue='SEX')
```
